```python
import pandas as pd

data = {
    "Date": [
        "2024-01-01 08:30", "2024-01-01 18:45", "2024-01-02 09:15",
        "2024-01-02 22:10", "2024-01-03 07:50", "2024-01-03 19:30"
    ],
    "Weather_Condition": [
        "Clear", "Rain", "Fog", "Rain", "Clear", "Clear"
    ],
    "Road_Condition": [
        "Dry", "Wet", "Wet", "Wet", "Dry", "Dry"
    ],
    "City": [
        "Delhi", "Mumbai", "Bengaluru", "Chennai", "Delhi", "Mumbai"
    ]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
df["Date"] = pd.to_datetime(df["Date"])
df["Hour"] = df["Date"].dt.hour

df
```

```python
df["Hour"].value_counts().sort_index().plot(kind="bar", title="Accidents by Hour")
```

```python
df["Weather_Condition"].value_counts().plot(kind="bar", title="Accidents by Weather")
```

```python
df["Road_Condition"].value_counts().plot(kind="bar", title="Accidents by Road Condition")
```
